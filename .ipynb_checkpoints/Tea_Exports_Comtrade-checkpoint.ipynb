# UN Comtrade Tea Exports (HS 0902) — End‑to‑End Downloader & MA(3) Top‑10 Panels

This notebook fetches **annual export values in USD** for **HS 0902 (Tea)** using the official `comtradeapicall` package, writes a clean CSV (`tea_exports_comtrade.csv`), and builds **3‑year moving‑average** snapshots for your lecture slide (Top‑10 exporters by selected years).

```python
import time
import pandas as pd
from typing import Iterable, Optional, Union

import comtradeapicall

def _chunks(seq: Iterable, size: int):
    seq = list(seq)
    for i in range(0, len(seq), size):
        yield seq[i:i+size]

def download_tea_exports(
    subscription_key: Optional[str] = None,
    years: Union[str, Iterable[int]] = range(1962, 2025),
    reporters: str = "all",  # "all" or comma-separated codes/ISO3
    partner: str = "0",      # 0 = World
    throttle_s: float = 1.2,  # sleep between calls
    use_tariffline: bool = False,
    proxy_url: Optional[str] = None,
) -> pd.DataFrame:
    """
    Download annual EXPORT values (USD) for HS 0902 via `comtradeapicall`.
    Returns tidy DataFrame: reporter, year, value_usd
    """
    if isinstance(years, str):
        periods = [years]
    else:
        years = list(map(int, years))
        blocks = list(_chunks(years, 10))
        periods = [f"{blk[0]}:{blk[-1]}" for blk in blocks]

    cmd_hs = "0902"
    frames = []
    for period in periods:
        if subscription_key:
            if use_tariffline:
                df = comtradeapicall.getTarifflineData(
                    subscription_key,
                    typeCode="C", freqCode="A", clCode="HS",
                    period=period, reporterCode=reporters, cmdCode=cmd_hs,
                    flowCode="X", partnerCode=partner,
                    maxRecords=500000, format_output="JSON",
                    includeDesc=True, proxy_url=proxy_url
                )
            else:
                df = comtradeapicall.getFinalData(
                    subscription_key,
                    typeCode="C", freqCode="A", clCode="HS",
                    period=period, reporterCode=reporters, cmdCode=cmd_hs,
                    flowCode="X", partnerCode=partner,
                    maxRecords=500000, format_output="JSON",
                    includeDesc=True, proxy_url=proxy_url
                )
        else:
            if use_tariffline:
                df = comtradeapicall.previewTarifflineData(
                    typeCode="C", freqCode="A", clCode="HS",
                    period=period, reporterCode=reporters, cmdCode=cmd_hs,
                    flowCode="X", partnerCode=partner,
                    maxRecords=20000, format_output="JSON",
                    countOnly=None, includeDesc=True, proxy_url=proxy_url
                )
            else:
                df = comtradeapicall.previewFinalData(
                    typeCode="C", freqCode="A", clCode="HS",
                    period=period, reporterCode=reporters, cmdCode=cmd_hs,
                    flowCode="X", partnerCode=partner,
                    maxRecords=20000, format_output="JSON",
                    aggregateBy=None, breakdownMode="classic",
                    countOnly=None, includeDesc=True, proxy_url=proxy_url
                )

        if df is None or len(df) == 0:
            time.sleep(throttle_s)
            continue

        cols = {c.lower(): c for c in df.columns}
        keep = {
            "year": df[cols.get("period")],
            "reporter": df[cols.get("reporterdesc", cols.get("reporter", cols.get("reportercode")))],
            "cmd": df[cols.get("cmdcode")],
            "flow": df[cols.get("flowcode")],
        }
        tv = None
        for k in ["primaryValue","TradeValue","tradeValue","primary_value","Value"]:
            if k in df.columns:
                tv = df[k]; break
        if tv is None:
            tv = df[cols.get("value")]
        keep["value_usd"] = tv

        part = pd.DataFrame(keep)
        part = part[(part["cmd"] == cmd_hs) & (part["flow"].astype(str).str.upper()=="X")][["reporter","year","value_usd"]]
        frames.append(part)
        time.sleep(throttle_s)

    out = pd.concat(frames, ignore_index=True).dropna(subset=["value_usd"])
    out = out.groupby(["reporter","year"], as_index=False)["value_usd"].sum()
    return out

def save_tea_exports_csv(path_csv: str, subscription_key=None, years=range(1962, 2025), reporters="all", partner="0", proxy_url=None) -> str:
    df = download_tea_exports(subscription_key, years, reporters, partner, proxy_url=proxy_url)
    df = df.rename(columns={"value_usd":"trade_value_usd"})
    df.to_csv(path_csv, index=False)
    return path_csv
```

## 2) Build 3‑year moving average (MA3) and save snapshot table
This prepares the data you use for Slide 5 (top‑10 exporters by chosen years).

```python
import numpy as np
import pandas as pd

trade = pd.read_csv(OUT_CSV)
trade = trade.sort_values(["reporter","year"])  

# Compute MA(3)
trade["ma3"] = (trade
                .groupby("reporter")["trade_value_usd"]
                .transform(lambda s: s.rolling(3, min_periods=3).mean()))

# Pick snapshot years (like your slide grid)
snap_years = list(range(1962, (int(trade["year"].max())//5)*5 + 2, 5))  # 1962,67,...
snap_years = [y for y in snap_years if y <= trade["year"].max()]

records = []
for y in snap_years:
    snap = trade[trade["year"]==y].dropna(subset=["ma3"]).nlargest(10, "ma3")
    for _, r in snap.iterrows():
        records.append({"year": y, "reporter": r["reporter"], "ma3_usd": r["ma3"]})

snap_df = pd.DataFrame(records)
snap_df.to_csv("tea_exports_top10_ma3_snapshots.csv", index=False)
snap_df.head()
```

## 3) Plot (optional) — one snapshot example

```python
import matplotlib.pyplot as plt

y = snap_years[-1]
ex = snap_df[snap_df["year"]==y].sort_values("ma3_usd")
plt.figure(figsize=(8,6))
plt.barh(ex["reporter"], ex["ma3_usd"])  # no style/colors set per your constraints
plt.title(f"Top Tea Exporters in {y} — MA(3) USD")
plt.xlabel("USD")
plt.tight_layout()
plt.savefig("top10_exporters_ma3_example.png", dpi=200)
print("Saved figure: top10_exporters_ma3_example.png")
```
